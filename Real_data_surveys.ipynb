### Users with a lot of distinct places AND a lot of records

I should be able to test with them

```python
users_distinct_place = df_ambiance[['user', 'place_type']].groupby('user').agg(['count']).sort_values(('place_type', 'count'), ascending=False).reset_index()['user'].tolist()[:10]
users_records = df_ambiance.groupby('user').size().reset_index(name='counts').sort_values('counts', ascending=False)['user'].tolist()[:10]
best_users = set(users_distinct_place) & set(users_records)
best_users
```
